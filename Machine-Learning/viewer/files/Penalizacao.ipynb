```python
from sklearn.linear_model import Ridge, LinearRegression, Lasso
from sklearn.model_selection import train_test_split
import mglearn
import plotly.express as px
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
X, y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)
ridge = Ridge().fit(X_train, y_train)
ridge10 = Ridge(alpha=10).fit(X_train, y_train)
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)
```

```python
df = (pd.DataFrame([np.hstack([lr.intercept_, lr.coef_]),
                    np.hstack([ridge.intercept_, ridge.coef_]),
                    np.hstack([ridge10.intercept_, ridge10.coef_]),
                    np.hstack([ridge01.intercept_, ridge01.coef_])]).
      T.
      rename(columns={0: 'logistica',
                      1:'ridge_1',
                      2:'ridge_10',
                      3: 'ridge_0.1'}))
```

```python
df
```

```python
px.scatter(df)
```

```python
print(f'MSE Regressao Logistica: {mean_squared_error(y_test, lr.predict(X_test))}')
print(f'MSE Regressao Ridge Alpha = 1: {mean_squared_error(y_test, ridge.predict(X_test))}')
print(f'MSE Regressao Ridge Alpha = 10: {mean_squared_error(y_test, ridge10.predict(X_test))}')
print(f'MSE Regressao Ridge Alpha = 0.1: {mean_squared_error(y_test, ridge01.predict(X_test))}')
```

```python
lasso = Lasso().fit(X_train, y_train)
lasso10 = Lasso(alpha=10).fit(X_train, y_train)
lasso01 = Lasso(alpha=0.1).fit(X_train, y_train)
```

```python
df = (pd.DataFrame([np.hstack([lr.intercept_, lr.coef_]),
                    np.hstack([lasso.intercept_, lasso.coef_]),
                    np.hstack([lasso10.intercept_, lasso10.coef_]),
                    np.hstack([lasso01.intercept_, lasso01.coef_])]).
      T.
      rename(columns={0: 'logistica',
                      1:'lasso_1',
                      2:'lasso_10',
                      3: 'lasso_0.1'}))
```

```python
df
```

```python
px.scatter(df)
```

```python
print(f'MSE Regressao Logistica: {mean_squared_error(y_test, lr.predict(X_test))}')
print(f'MSE Regressao Lasso Alpha = 1: {mean_squared_error(y_test, lasso.predict(X_test))}')
print(f'MSE Regressao Lasso Alpha = 10: {mean_squared_error(y_test, lasso10.predict(X_test))}')
print(f'MSE Regressao Lasso Alpha = 0.1: {mean_squared_error(y_test, lasso01.predict(X_test))}')
```
