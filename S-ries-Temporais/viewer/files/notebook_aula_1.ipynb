# **Desafio**
Seu time recebe a demanda de desenvolver um modelo preditivo.
As primeiras etapas do ciclo Crisp DM foram concluídas e o
o modelo foi encaminhado para produtização. Passados 2 meses, o engenheiro de ML responsável pela execução informou que o modelo retornara erro. Ao checar
o processo de atualização das bases de dados, verificou-se que haviam dados ausentes. O que pode ter ocorrido?

\\
**Resposta**
1. Mudança na Qualidade ou Disponibilidade dos Dados
Falha no ETL (Extract, Transform, Load): O processo que carrega os dados para alimentar o modelo pode ter falhado, causando ausência de informações críticas.
Inconsistência na Origem dos Dados: Alterações no formato ou estrutura dos dados na base de origem podem ter provocado omissões ou incompatibilidades.
2. Dados Inesperadamente Nulos
Mudanças em Sistemas Fonte: Sistemas que fornecem os dados podem ter mudado seus campos ou lógicas, resultando em valores nulos ou ausentes em atributos antes preenchidos.
Campos Obrigatórios Não Populados: Se os dados ausentes eram essenciais para o modelo, o processo de inferência falhou ao não encontrar as entradas esperadas.
3. Falta de Validação no Pipeline
Ausência de Monitoramento de Dados: Um mecanismo de validação ou alerta para dados ausentes não foi implementado, permitindo que dados inconsistentes chegassem ao modelo.
Data Drift ou Esquema Drift: Mudanças na distribuição dos dados (data drift) ou alterações no esquema das bases (schema drift) podem ter gerado inconsistências.
4. Problemas no Processo de Atualização
Bases Não Atualizadas: Se o modelo dependia de dados novos (exemplo: dados de tempo real ou mensais) e a atualização não ocorreu, os dados ausentes poderiam causar erros na inferência.
Erros no Agendamento: Jobs responsáveis por atualizar a base podem ter falhado ou não terem sido executados corretamente.
5. Imputação de Dados Não Considerada
Falta de Estratégia de Preenchimento: Se o modelo não foi projetado para lidar com dados ausentes (ex.: imputação, exclusão ou outros métodos), ele pode retornar erro ao encontrar entradas nulas.

Dataset: beer   
Descrição: Produção mensal de cerveja na Austrália

```python
fig, ax = plt.subplots(figsize = (12,6))
df_full_beer.plot(ax = ax, color = 'gray', title = 'Produção mensal de cerveja na Austrália', legend = '');
ax.set_ylabel('Unidades', labelpad = 10);
```

* Dados da População da Austrália -> Mês para Ano

```python
df_pop_aust = pd.read_csv('pop_aust.csv')
df_pop_aust['ano'] = pd.DatetimeIndex(df_pop_aust['ano'].astype(str))
df_pop_aust = df_pop_aust.set_index('ano')
df_pop_aust.index.name = None
df_pop_aust = df_pop_aust.div(1e6)
```

* Produção Anual de Cerveja Per Capita

```python
df_beer_per_capita = df_full_beer.resample('YS').sum().join(df_pop_aust).eval('beer_per_capita = beer_production / populacao')
df_beer_per_capita = df_beer_per_capita['beer_per_capita'].dropna()

fig, ax = plt.subplots(figsize = (12,6))
df_beer_per_capita.plot(ax = ax, color = 'gray', title = 'Produção anual de cerveja na Austrália - Per Capita', legend = '');
ax.set_ylabel('Unidades', labelpad = 10);
```

A respeito da série com histórico completo:

* A série possui indicativo de tendência? **R: Sim.**
    * A tendência aparenta ser determinística? **Não**.
* Alguma transformação pode ser útil neste caso? Produção per Capita
    * Se sim, qual? **Divisão da Produção pelo tamanho da população.**
* Podemos obter algum insight a respeito? **Que no ano de 1970 ate 1980 teve um grande aumento. Ou devido ao aumento da população consequêntemente aumentou o consumo de alcool.**

```python
# prompt: agora crie varios graficos de linhas para cada ano com as vedas de varejo juntando em um grafico e o label esta quebrando no grafico tem como arrumar

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# ... (Your existing code for data loading and preprocessing)

# Create the pivoted DataFrame
df_varejo_pivotado = df_varejo.pivot_table(index=df_varejo.index.year, columns=df_varejo.index.month, values='valor')

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))

for year in df_varejo_pivotado.index:
    ax.plot(df_varejo_pivotado.loc[year], label=str(year))

# Customize the x-axis ticks to show month numbers
ax.xaxis.set_major_locator(ticker.FixedLocator(range(1, 13)))
ax.xaxis.set_major_formatter(ticker.FixedFormatter([str(month) for month in range(1, 13)]))


ax.set_title('Índice de volume de vendas no varejo - Total - Brasil')
ax.set_xlabel('Mês')
ax.set_ylabel('Valor')
ax.legend(title="Ano", bbox_to_anchor=(1.05, 1), loc='upper left') # Move the legend outside

plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

* Os gráficos a seguir podem ser úteis para observar os padrões sazonais em uma série temporal.

* Execute o gráfico com a matriz transposta. Obs: Meses em Colunas e Anos em linhas.

```python
# Plotando o gráfico
fig, ax = plt.subplots(figsize=(12, 6))

for month in df_varejo_pivotado.index:
    ax.plot(df_varejo_pivotado.columns, df_varejo_pivotado.loc[month], label=f'Mês {month}')

# Personalizar os eixos
ax.set_title('Índice de Volume de Vendas no Varejo - Total - Brasil', fontsize=14)
ax.set_xlabel('Ano', fontsize=12)
ax.set_ylabel('Valor', fontsize=12)

# Configurar ticks no eixo X para os anos
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

# Adicionar legenda
ax.legend(title="Mês", bbox_to_anchor=(1.05, 1), loc='upper left')  # Mover legenda para fora do gráfico

# Ajustar layout
plt.tight_layout()
plt.show()
```

```python
# Configurar subplots
fig, axes = plt.subplots(nrows=1, ncols=12, figsize=(24, 6), sharey=True)  # 12 subplots (1 por mês)

# Plotar cada mês em um gráfico separado
for i, month in enumerate(df_varejo_pivotado.columns):
    ax = axes[i]
    ax.plot(df_varejo_pivotado.index, df_varejo_pivotado[month], label=f'Mês {month}')

    # Adicionar linha da média
    media_mes = df_varejo_pivotado[month].mean()
    ax.axhline(y=media_mes, color='red', linestyle='--', label='Média')

    # Configurações do subplot
    ax.set_title(f'Mês {month}', fontsize=10)
    ax.tick_params(axis='x', rotation=45)  # Rotacionar os rótulos do eixo X
    ax.legend(fontsize=8)  # Adicionar legenda para cada subplot

# Ajustar layout
plt.tight_layout()
plt.show()
```
