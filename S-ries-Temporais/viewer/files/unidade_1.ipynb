Dataset: beer   
Descrição: Produção mensal de cerveja na Austrália

```python
df_full_beer = pd.read_csv('full_bear.csv', index_col = 0)
titulo = 'Produção mensal de cerveja na Austrália'
titulo_eixo_y = 'Unidades'

df_full_beer.columns = ['beer_production']
df_full_beer.index.name = None

# Conferir os dtypes
print(f"Coluna: {df_full_beer['beer_production'].dtype}")
print(f"Índice: {df_full_beer.index.dtype}")

# Ajuste dos dtypes
df_full_beer['beer_production'] = df_full_beer['beer_production'].astype(float)
df_full_beer.index = pd.DatetimeIndex(df_full_beer.index)

fig, ax = plt.subplots(figsize = (12,6))
df_full_beer.loc['1991': '1996'].plot(ax = ax, color = 'gray', title = titulo, legend = '');
ax.set_ylabel(titulo_eixo_y, labelpad = 10);
```

* Dados da População da Austrália -> Mês para Ano

```python
df_pop_aust = pd.read_csv('pop_aust.csv')
df_pop_aust['ano'] = pd.DatetimeIndex(df_pop_aust['ano'].astype(str))
df_pop_aust = df_pop_aust.set_index('ano')
df_pop_aust.index.name = None
df_pop_aust = df_pop_aust.div(1e6)
```

* Produção Anual de Cerveja Per Capita

```python
df_beer_per_capita = df_full_beer.resample('YS').sum().join(df_pop_aust).eval('beer_per_capita = beer_production / populacao')
df_beer_per_capita = df_beer_per_capita['beer_per_capita'].dropna()

fig, ax = plt.subplots(figsize = (12,6))
df_beer_per_capita.plot(ax = ax, color = 'gray', title = 'Produção anual de cerveja na Austrália - Per Capita', legend = '');
ax.set_ylabel('Unidades', labelpad = 10);
```

A respeito da série com histórico completo:

* A série possui indicativo de tendência? **R: Sim.**
    * A tendência aparenta ser determinística? **Não**.
* Alguma transformação pode ser útil neste caso? Produção per Capita
    * Se sim, qual? **Divisão da Produção pelo tamanho da população.**
