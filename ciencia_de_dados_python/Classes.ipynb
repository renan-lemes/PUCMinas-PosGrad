## Método static e class

```python
class Realiza_operacao:
  def __init__(self, constante):
    self.c = constante

  def adicao_constante(self, x):
    return x+self.c

  @staticmethod
  def adicao_numeros(x, y):
    return x+y
```

```python
import datetime

class ProcessDate():
  query = """
  select *
  from table
  where data >= '{}' and data <= '{}'
  """

  def __init__(self, initial_date, final_date):
    self.initial_date = initial_date
    self.final_date = final_date


  def format_query(self):
    return self.query.format(self.initial_date, self.final_date)

  @classmethod
  def set_final_date(cls, initial_date, time):
    return cls(initial_date, initial_date+datetime.timedelta(time))
```

```python
pd = ProcessDate(datetime.date(2023, 3, 1), datetime.date(2023, 4, 1))
pd.format_query()
```

```python
pd.set_final_date(datetime.date(2023, 4, 1), 15)
```

```python
pd.initial_date
```

```python
pd2 = pd.set_final_date(datetime.date(2023, 4, 1), 15)
pd2.format_query()
```

```python
pd2.initial_date
```

```python
ProcessDate.set_final_date(datetime.date(2023, 4, 1), 15).format_query()
```

```python
import time

from functools import wraps

def timed(func):

    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f'Tempo {end-start} s')
        return result

    return wrapper
```

```python
@timed
def soma_valores(lista):
  soma = 0
  for k in lista:
    soma += k
  return soma
```

```python
soma_valores([1, 2, 5, 10])
```

```python
class Preprocessamento():
  c = 0.5

  def __init__(self, lista_dados):
    self.dados = lista_dados

  def adiciona_constante(self):
    return [value + self.c for value in self.dados]

  def filtrar_dados(self):
    pass
```

```python
class Preprocessamento_limite(Preprocessamento):
  def __init__(self, lista_dados, limiar):
    super().__init__(lista_dados)
    self.limiar = limiar

  def filtrar_dados(self, dados=None):
    if not dados:
      dados = self.dados
    return [value for value in dados if value>self.limiar]

  def filtro_constante(self):
    dados_const = self.adiciona_constante()
    return self.filtrar_dados(dados_const)
```

```python
x1 = Preprocessamento([0.2, 1, 1.5])
x1.adiciona_constante()
```

```python
x1.filtrar_dados()
```

```python
x2 = Preprocessamento_limite([0.2, 1, 1.5], 0.6)
```

```python
x2.filtrar_dados()
```

```python
x2.filtro_constante()
```
