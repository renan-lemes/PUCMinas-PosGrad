## Mais de uma condição

```python
valor = 100
canal = 'site'

if valor >= 100:
  print('Primeiro caso')
elif canal == 'site':
  print('Segundo caso')
else:
  print('Terceiro caso')
```

## Utilização do and

```python
receita = 500
canal = 'site'
if (receita > 200) & (canal=='site'):
  print('Temos o caso 1')
else:
  print('Temos o caso 2')
```

```python
receita = 500
canal = 'site'
if receita > 200 and canal=='site':
  print('Temos o caso 1')
else:
  print('Temos o caso 2')
```

```python
logica = receita > 200
logica &= canal == 'site'
if logica:
  print('Temos o caso 1')
else:
  print('Temos o caso 2')
```

## Utilizando or

```python
receita = 500
canal = 'site'
if (receita > 200) | (canal=='site'):
  print('Temos o caso 1')
else:
  print('Temos o caso 2')
```

```python
receita = 500
canal = 'site'
if receita > 200 or canal=='site':
  print('Temos o caso 1')
else:
  print('Temos o caso 2')
```

```python
logica = receita > 200
logica |= canal == 'site'
if logica:
  print('Temos o caso 1')
else:
  print('Temos o caso 2')
```

## Validando uma lógica

```python
a = True
b = False
print(f'Lógica e: {a & b}')
print(f'Lógica ou: {a | b}')
```

```python
a = False
b = False
print(f'a ou b: {a or b}')
print(f'Não é (a ou b): {not (a  or b)}')
print(f'Não é a e não é b: {(not a) and (not b)}')
```

```python
a = True
b = False
print(f'a e b: {a and b}')
print(f'Não é (a e b): {not (a  and b)}')
print(f'Não é a ou não é b: {(not a) or (not b)}')
```

## Multiplas condições

```python
receita = 100
tarifa = 'mensal'
canal = 'loja'

if tarifa == 'diaria':
  if canal in ['app', 'site']:
    receita = receita * 0.9
  else:
    receita *= 0.7
else:
  receita *= 0.8
print(receita)
```

## Criando variável com if

```python
venda = False
oferta = 100

receita = oferta if venda else 0
print(receita)
```
