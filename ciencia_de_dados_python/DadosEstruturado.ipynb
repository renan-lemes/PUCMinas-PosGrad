# Dados estruturados

Anteriormente vimos array que são homogêneos no tipo de dados, aqui iremos apresentar array com dados heterogêneos.

```python
import numpy as np
```

```python
dados = np.zeros(3, dtype={'names': ('nome', 'idade', 'notas'), 'formats': ('U10', 'i4', 'f8')})
dados
```

```python
dados['nome'] = ['Gabriel', 'Ana', 'Amanda']
dados['idade'] = [30, 25, 27]
dados['notas'] = [6, 4, 3.5]
dados
```

```python
dados['nome']
```

```python
dados[0]
```

```python
dados.dtype
```

```python
dados.dtype.names
```

```python
dados.dtype.fields
```

```python
dados = np.array([('Gabriel', 30, 6), 
                  ('Ana', 25, 4),
                  ('Amanda', 27, 3.5)], dtype={'names': ('nome', 'idade', 'notas'), 'formats': ('U10', 'i4', 'f8')})
```

```python
dados
```

```python
dados[dados['idade']<= 27]
```
