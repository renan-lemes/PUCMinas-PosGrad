## Operações com datas

```python
datetime.date(2023, 4 ,5) - datetime.date(2023, 4, 1)
```

```python
datetime.date(2023, 4 ,5) + datetime.timedelta(-15)
```

```python
datetime.date(2023, 4 ,5) + datetime.timedelta(2)
```

```python
datetime.datetime(2023, 4 ,5, 12) + datetime.timedelta(hours=2)
```

## Trabalhando com data e string

* Podemos converter data para string utilizando strftime
* Podemos converter string para data utilizando strptime


String para formato
* %a -> abreviação do dia da semana
* %A -> dia da semana
* %w -> dia da semana númerico
* %d -> dia do mês
* %b -> abreviatura nome do mês
* %B -> nome do mês
* %m -> mês
* %y -> ano (00, 01)
* %Y -> ano (2000, 2001)
* %H -> hora (24 horas)
* %I -> hora (12 horas)
* %p - AM ou PM
* %M -> minuto
* %S -> segundos
* %f -> microsegundos

```python
now.strftime('%H:%M:%S')
```

```python
now.strftime('%H:%M:%S %a')
```

```python
datetime.datetime.strptime('31 mar 00', '%d %b %y')
```
