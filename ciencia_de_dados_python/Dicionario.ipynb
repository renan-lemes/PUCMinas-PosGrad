# Dicionário
* Dicionários são criados com {}, sempre os valores tem que estar no formato chave:valor
* Acessamos um valor no dicionário com [chave]

```python
dict_exemplo = {'chave_1': 1, 'chave_2': 2}
dict_exemplo['chave_1']
```

```python
dict_exemplo['chave_3']
```

```python
dict_exemplo = {'chave_1': 1, 'chave_2'}
```

```python
lista_exemplo = [1, 3, 4]
dict_exemplo = {'chave_1': 1, 'chave_2': 2, 'chave_3': lista_exemplo}
lista_exemplo.append(5)
print(dict_exemplo)

dict_2 = {0: dict_exemplo}

dict_exemplo['chave_1'] = 3
print(dict_exemplo)

print(dict_2)
```

## As operações que podem ser realizadas com dicionário:
* update - acrescenta um valor no dicionário
* pop - remove um valor baseado na chave
* keys - retorna as chaves do dicionário
* values - retorna os valores do dicionário
* items - retorna (chave, valor) do dicionário

```python
dict_exemplo.update({'a': 'b'})
print(dict_exemplo)
```

```python
dict_exemplo.update({'chave_3': 2})
print(dict_exemplo)
```

```python
dict_exemplo.update({'chave_4': 'b'})
print(dict_exemplo)
```

```python
dict_exemplo.pop('a')
print(dict_exemplo)
```

```python
dict_exemplo = {'chave_1': 1, 'chave_2': 2, 'chave_3': lista_exemplo}
dict_exemplo.keys()
```

```python
dict_exemplo.values()
```

```python
dict_exemplo.items()
```

## Utilização de get e setdefault
Uma operação muito útil de dicionário é a função com get, com ela acessamos um valor do dicionário através de uma chave. Se essa chave não existe podemos retornar um valor default.


Uma outra operação muito útil neste sentido é a setdefault, com ela podemos adicionar uma informação no dicionário com um valor default. Observação só será adicionado caso a chave já não exista.


```python
dict_exemplo = {'chave_1': 1, 'chave_2': 2, 'chave_3': lista_exemplo}
dict_exemplo.get('chave_1', 'Não existe a chave')
```

```python
dict_exemplo.get('chave_4', 'Não existe a chave')
```

```python
dict_exemplo.setdefault('chave_4')
```

```python
print(dict_exemplo)
```

```python
dict_exemplo.setdefault('chave_1')
```

```python
dict_exemplo.setdefault('chave_5', 0)
```

```python
print(dict_exemplo)
```

## Criação de lista com chaves já determinadas

```python
dicionario = dict.fromkeys(['Gabriel', 'Amanda'])
print(dicionario)
```

```python
dicionario = dict.fromkeys(['Gabriel', 'Amanda'], 1)
print(dicionario)
```

```python
dicionario = dict.fromkeys(['Gabriel', 'Amanda'], [1,2])
print(dicionario)
```
