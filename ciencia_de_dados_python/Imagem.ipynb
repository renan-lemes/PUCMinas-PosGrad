# Alterando imagems com Numpy

```python
from PIL import Image
import numpy as np
```

```python
im = np.array(Image.open('cat.jpeg'))
```

```python
im.shape #altura, largura, cor (RGB)
```

## Alterando a cor

```python
im[100, 150] # (R, G, B)
```

```python
##mudando a cor

im_R = im.copy()
im_R[:, :, (1, 2)] = 0
im_G = im.copy()
im_G[:, :, (0, 2)] = 0
im_B = im.copy()
im_B[:, :, (0,1)] = 0

Image.fromarray(im_R)
```

```python
Image.fromarray(im_G)
```

```python
Image.fromarray(im_B)
```

```python
Image.fromarray(255 - im) #255 é o valor máximo
```

```python
Image.fromarray(im // 16 * 16) #redução de cor
```

```python
Image.fromarray(im // 32 * 32)
```

```python
Image.fromarray(im // 64 * 64)
```

## Cortando a imagem

```python
Image.fromarray(im[50:150, 128:280])
```

## Rotacionando a imagem

```python
np.rot(im)
```

```python
Image.fromarray(np.rot90(im))
```

```python
Image.fromarray(np.rot90(im, 2))
```

## Mesclando imagens

```python
im2 = np.array((Image.open('dog.jpg')).resize(im.shape[1::-1]))
```

```python
combine = (im * 0.6 + im2 * 0.4).astype(np.uint8)
Image.fromarray(combine)
```

## Adicionando máscara

```python
im_0 = im[15:, 10:-17, :]
```

```python
Image.fromarray(im_0)
```

```python
bx = int(im.shape[0] * 0.1)
by = int(im.shape[1] * 0.2)
```

```python
mask = np.zeros(im.shape)
mask[bx:-bx, by:-by, :] = 1
```

```python
masked_image = im * mask
```

```python
Image.fromarray(masked_image.astype('uint8'))
```
