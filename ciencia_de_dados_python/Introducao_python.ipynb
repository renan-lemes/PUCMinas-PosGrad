### Operações básicas com string:

* split - quebrar a string utilizando um caractere como referência
* startswith, endswith - verifica se a string começa(termina) com um caractere específico
* lower, upper, capitalize - funções para transformar caracteres em minúsculo e maiúsculo.
* count - contagem da frequência de um caractere na string
* find - identifica em que posição começa um caractere específico
* format - formata a string colocando variáveis
* encode - codifica a string
* islower, isupper, isnumeric, isalpha - válida se a string é minúscula, maiúscula, numérica ou faz parte do alfabeto.
* replace - substitui um caractere por outro
* strip - remove espaço no início e fim da string

```python
"String para ser quebrada".split()
```

```python
"String para ser quebrada".split('a')
```

```python
"String para ser quebrada".split('c')
```

```python
print('Teste começa'.startswith('T'))
print('Teste começa'.startswith('t'))
print('Teste começa'.startswith('Teste'))
```

```python
print('Teste encerra'.endswith('a'))
print('Teste encerra'.endswith('A'))
print('Teste encerra '.endswith(' '))
```

```python
print('Minúsculo Para a frase'.lower())
print('Maiúsculo Para a frase'.upper())
print('maiúsculo Para o inicio'.capitalize())
```

```python
print('abc'.find('a'))
print('abc'.find('b'))
print('abc'.find('ab'))
print('abc'.find('bc'))
print('abc'.find('d'))
print('abc'.find('e'))
print('abca'.find('a'))
```

```python
print('Quantos as temos nessa frase?'.count('a'))
print('Quantos as temos nessa frase?'.count('b'))
```
