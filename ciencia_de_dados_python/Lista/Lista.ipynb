# Lista

```python
lista_vazia = []
print(lista_vazia)

outra_vazia = list()
print(outra_vazia)

primeira_lista = [1, 2, 3, 4, 5]
print(f'Exemplo de lista{primeira_lista}')
segunda_lista = [1, 2, 'a']
print('Exemplo de lista {}'.format(segunda_lista))
terceira_lista = [primeira_lista, segunda_lista]
terceira_lista
```

## Acessando informações na lista

Assim como com strings é possível acessar a informação em um ponto específico da lista utilizando []

**Obs.**: Uma diferença da lista e string é que podemos alterar o valor em uma posição da lista

```python
len(primeira_lista)
```

```python
print(f'Valor na posição 0: {primeira_lista[0]}')
print(f'Valor na posição 2: {segunda_lista[2]}')
print(f'Acessando os valores da posição 0 a 2: {primeira_lista[0:2]}')
print(f'Acessando os valores da posição 0 a 4, com espaço de 2 :{primeira_lista[0:4:2]}')
print(f'Acessando o último valor da lista: {segunda_lista[:-1]}')
print(f'Acessando o primeiro valor da lista: {terceira_lista[1]}')
print(terceira_lista[2])
```

```python
primeira_lista[1] = 5
primeira_lista
```

```python
terceira_lista
```

## Adicionando informação a lista

* Podemos usar a operação + para juntar mais de uma lista em uma única; isso não altera as listas originais.
* Caso desejamos adicionar apenas uma informação podemos usar o append
* Caso desejamos adicionar mais de uma informação podemos utilizar extend, essa operação é similar ao + porém altera a própria lista.
* Caso desejamos adicionar informação em uma posição específica utilizar insert

```python
primeira_lista + segunda_lista
```

```python
primeira_lista.append('a')
primeira_lista
```

```python
primeira_lista.append(segunda_lista)
primeira_lista
```

```python
lista_exemplo = [1, 2]
lista_exemplo.extend(segunda_lista)
lista_exemplo
```

```python
lista_1 = [1, 2]
lista_1 += [1, 2, 'a']
print(lista_1)
```

```python
lista_exemplo = [1, 2]
print(lista_exemplo)
lista_exemplo.insert(1, 5)
print(f'Inserindo o valor 5 na posição 1 : {lista_exemplo}')
print(f'Valor na posição 1: {lista_exemplo[1]}')
lista_exemplo.insert(20, 5)
print(f'Inserindo o valor 20 na posição 5: {lista_exemplo}')
lista_exemplo.insert(-2, 'a')
print(f'Adicionando o valor a na penúltima posição: {lista_exemplo}')
print(lista_exemplo[-2])
```

## Removendo informação da lista
* remove - remove um valor da lista
* pop - remove um valor baseado na posição

```python
lista_exemplo = [3, 2, 3, 4]
lista_exemplo.remove(3)
print(lista_exemplo)
lista_exemplo.remove(5)
```

```python
lista_exemplo.pop(1)
print(lista_exemplo)
```

## Invertendo uma lista

Em algumas operações é interessante inverter a ordem de uma lista por exemplo para remover valores baseado em uma condição lógica.

```python
lista_exemplo = [1, 2, 3]
lista_exemplo[::-1] #não altera a lista original
```

```python
lista_exemplo.reverse()
lista_exemplo
```

## Outras operações

* count - contar a frequência de um valor
* max, min - obter o máximo e mínimo de uma lista
* sum - realiza a soma dos valores da lista
* index - índice em qual posição se encontra um valor
* sort - ordenar
* clear - limpar a lista

```python
lista_teste = [1, 2, 1, 3, 4, 5, 1]
lista_teste.count(1)
```

```python
max(lista_teste)
```

```python
min(lista_teste)
```

```python
sum(lista_teste)
```

```python
lista_teste.index(1)
```

```python
lista_teste.index(1, 1, 4) #definindo inicio e fim da busca
```

```python
valor_maximo = max(lista_teste)
lista_teste.index(valor_maximo)
```

```python
lista_teste.sort()
lista_teste
```

```python
lista_teste.clear()
lista_teste
```

```python
[1] * 5
```

## Funções lógicas com lista

```python
lista_exemplo = [1, 3, 4]
print(1 in lista_exemplo)
print(5 in lista_exemplo)
```

```python
print(5 not in lista_exemplo)
```

```python
lista_bool = [True, False, True]
print(any(lista_bool))
print(all(lista_bool))
```

```python
lista_false = [False, False]
print(any(lista_false))
lista_true = [True, True]
print(all(lista_true))
```

**Atenção**

Lista são elementos mutaveis

```python
lista_1 = [1, 2, 3]
lista_2 = lista_1
lista_1.append(4)
print(lista_2)
lista_2.remove(3)
print(lista_1)
```

```python
lista_1 = ['oi', 'tudo', 'bem']
lista_2 = [1, 2, lista_1]
print(lista_2)
lista_2[2] = 2
print(lista_1)
print(lista_2)
```

```python
lista_1 = ['oi', 'tudo', 'bem']
lista_2 = [1, 2, lista_1]
print(lista_2)
lista_1.append('?')
print(lista_2)
```

```python
lista_1 = [1, 2, 3]
lista_2 = lista_1.copy()
print(f'Lista 1: {lista_1}')
print(f'Lista 2: {lista_2}')
print('Adicionando 4  a lista 1')
lista_1.append(4)
print(f'Lista 1: {lista_1}')
print(f'Lista 2: {lista_2}')
print('Removendo 3 da lista 2')
lista_2.remove(3)
print(f'Lista 1: {lista_1}')
print(f'Lista 2: {lista_2}')
```
