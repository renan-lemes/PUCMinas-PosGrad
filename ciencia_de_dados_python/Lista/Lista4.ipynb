# Exercício 1


O dataset Titanic disponível no pacote seaborn contém dados dos passageiros do Titanic. Os campos presente no Dataset são:


1. survived: variável binária (1 se sobreviveu ou 0 contrario)
2. pclass: classe do ticket
3. sex: sexo do passageiro
4. age: idade em anos completos
5. sibsp: número de irmãos/cônjuges a bordo do navio
6. parch: número de pais/filhos a bordo do navio
7. fare: tarifa da passagem
8. embarked: porto de embarque (C=Cherbour, Q=Queenstown, S=Southampton)


Faça as análises a seguir:
1. Construa um boxplot com a distribuição da idade para os pacientes que sobreviveram e não sobreviveram. 
2. Construa um gráfico de barras que mostre o número de passageiros que sobreviveu e não sobreviveu por classe.
3. Construa um scatter plot mostrando a relação entre tarifa e idade; a cor de cada ponto deve ser referente a classe.

Dicas para as análises:

2. Para construir o gráfico construa uma tabela intermediária onde os índices são as classes e as colunas sobreviveu ou não.

```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
## Analise 1

import matplotlib.pyplot as plt ## importei mais para fazer os dois em um subplot

df_survived = titanic[titanic["survived"] == 1] ## separei em dois dfs de sobreviventes e não sobreviventes
df_not_survived = titanic[titanic["survived"] == 0]

fig, axes = plt.subplots(1, 2, figsize=(15, 8)) ## Crio as figuras e seus axs

sns.boxplot(x="pclass",y="age",  data=df_survived, ax=axes[0], palette="Set2") ## crio o primeiro para os que sobreviverão
axes[0].set_title('Boxplot da idade das pessoas que sobreviverão') ## Coloco o titulo para a figura 0

sns.boxenplot(x="pclass", y="age", data=df_not_survived, ax=axes[1], palette="Set2") ## crio o a segunda figura para o df com as pessoas que não sobreviverão
axes[1].set_title('Boxplot da idade das pessoas que não sobreviverão') ## aqui crio a legenda para o a figura 1

plt.show()
```

```python
## Analise 1 se for analisar para classe mulheres e homens
 
df_survived = titanic[titanic["survived"] == 1] 
df_not_survived = titanic[titanic["survived"] == 0]

fig, axes = plt.subplots(1, 2, figsize=(15, 8)) 

sns.boxplot(x="sex",y="age",  data=df_survived, ax=axes[0], palette="Set2") 
axes[0].set_title('Boxplot da idade das pessoas que sobreviverão') 

sns.boxenplot(x="sex", y="age", data=df_not_survived, ax=axes[1], palette="Set2") 
axes[1].set_title('Boxplot da idade das pessoas que não sobreviverão') 

plt.show()
```

```python
## Analise 2

## Cria a figura e os eixos 
fig, axes = plt.subplots(1, 2, figsize=(15, 8))

## criamos um countplots para contar a quantidade de sobreviventes
sns.countplot(x="pclass", hue="survived", data=df_survived, palette="coolwarm", ax=axes[0])
axes[0].set_title("Gráfico de Barras dos Sobreviventes")

## aqui criamos outro countplots para contar a quantidade de não sobreviventes
sns.countplot(x="pclass", hue="survived", data=df_not_survived, palette="coolwarm", ax=axes[1])
axes[1].set_title("Gráfico de Barras dos Não Sobreviventes")

## mostrar grafico
plt.show()
```

```python
## Analise 2 - porem analisando em relacao ao sexo 

fig, axes = plt.subplots(1, 2, figsize=(15, 8))

sns.countplot(x="sex", hue="survived", data=df_survived, palette="coolwarm", ax=axes[0])
axes[0].set_title("Gráfico de Barras dos Sobreviventes")

sns.countplot(x="sex", hue="survived", data=df_not_survived, palette="coolwarm", ax=axes[1])
axes[1].set_title("Gráfico de Barras dos Não Sobreviventes")

plt.show()
```

```python
## Analise 3

plt.figure(figsize=(15, 6)) ## Crio o tamanho da imagem

sns.scatterplot(x="fare", y="age", hue="pclass", data=titanic, palette="Set1") ## Escolhendo o grafico
plt.title("Scallter plot Tarifa e Idade") ##  colocando um titulo
plt.xlabel("Tarifa") ## colocando a descricao no eixo x
plt.ylabel("Idade") ## colocando a descricao no eixo y

plt.show()

```
