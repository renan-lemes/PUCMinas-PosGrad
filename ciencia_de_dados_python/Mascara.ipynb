# Utilizando booleano em Arrays

```python
import numpy as np
```

```python
x = np.arange(10)
```

```python
x < 5
```

```python
x >= 5
```

```python
(3 * x) == (x ** 2)
```

## Operações com booleanos

```python
np.sum(x < 5)
```

```python
np.all(x < 5)
```

```python
np.any(x<5)
```

```python
x = np.arange(1, 10).reshape((3,3)).T
```

```python
x
```

```python
np.sum(x<5, axis=1)
```

```python
np.sum(x<5, axis=0)
```

```python
np.any(x<5, axis=0)
```

```python
np.all(x<5, axis=0)
```

## Combinando lógica

```python
x = np.arange(10)

(x>5) & (x<8)
```

```python
(x>5) | (x<3)
```

```python
~((x>5)|(x<3))
```

```python
(x>5) ^ (x<8)
```

## Máscara

```python
x[x>=5]
```

```python
x[(x>5) | (x<3)]
```

## Igualdade de arrays

```python
a1 = np.array([1/3, 2/3, 3/3, 4/3])
a2 = np.array([2/6, 4/6, 6/6, 8/6])

a1 == a2
```

```python
np.all(a1 == a2)
```

```python
np.any(a1 == a2)
```

```python
a1 = np.array([1/3, 2/3, 3/3, 4/3])
a2 = np.array([0.33, 0.66, 1, 1.33])

a1 == a2
```

```python
np.all(a1 == a2)
```

```python
np.any(a1 == a2)
```

```python
np.isclose(a1, a2, atol=0.01)
```

```python
np.allclose(a1, a2, atol=0.01)
```

## Mudando valores em um array

```python
x = np.arange(10)
```

```python
np.place(x, x>5, [0, 1])
```

```python
np.where(x > 5, x, x*10)
```

```python
np.put(x, 0, 50)
```

```python
np.put(x, 12, 50)
```

```python
np.put(x, 12, 50, mode='clip')
```

```python
np.put(x, [0, 5], [50, 55])
```

```python
np.select(x, [x<3, x>5], [x, x-10], -100)
```

## Fancy Indexing

```python
x[[2, 4, 8]]
```

```python
ind = np.array([[2, 4], [8, 9]])
x[ind]
```

```python
x = np.arange(1, 13).reshape((3, 4))
x
```

```python
row = np.array([0, 1, 2])
col = np.array([2, 1, 3])
x[row, col]
```

```python
row[:, np.newaxis]
```

```python
x[row[:, np.newaxis], col]
```

```python
x[2, [2, 0, 1]]
```

```python
x[2:, [2, 0, 1]]
```

```python
mascara = np.array([False, True, False, True])
x[row[:, np.newaxis], mascara]
```
