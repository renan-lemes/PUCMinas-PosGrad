## RE

```python
import re
```

```python
padrao = r'(A|R|E)(d|e)?\w'
texto = 'As Reutilizações é muito importante na Escola'
caso = re.search(padrao, texto)
caso.start()
```

```python
caso.end()
```

```python
caso[0]
```

```python
for caso in re.finditer(padrao, texto):
  print(caso)
```

```python
re.sub(padrao, '', texto)
```

```python
re.split(padrao, texto)
```

```python
padrao = r'(A|R|E)(d|e)?(\w)'
```

```python
re.split(padrao, texto)
```
