# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
concat_data = np.hstack([iris['data'], iris['target'][:, np.newaxis]])
df_iris = pd.DataFrame(concat_data, 
                       columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target'])
tips = sns.load_dataset("tips")
df = sns.load_dataset("penguins")
```

## Scatterplot

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='target', size='petal_length', 
                palette='viridis',data=df_iris)
plt.legend(loc='upper left', bbox_to_anchor=(1.02, 1))
```

## Distribuição

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
sns.scatterplot(x=x, y=y, s=5, color=".15")
```

```python
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap="mako")
```

```python
sns.kdeplot(x=x, y=y, levels=5, color="b", linewidths=1)
```

```python
sns.scatterplot(x=x, y=y, s=5, color=".15")
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap="mako")
sns.kdeplot(x=x, y=y, levels=5, color="w", linewidths=1)
```

```python
sns.jointplot(
    data=df,
    x="bill_length_mm", y="bill_depth_mm", hue="species",
    kind="kde",
)
```

```python
graph = sns.PairGrid(df, diag_sharey=False)
graph.map_upper(sns.scatterplot, s=15)
graph.map_lower(sns.kdeplot)
```

## Heatmap

```python
from string import ascii_letters
rs = np.random.RandomState(33)
d = pd.DataFrame(data=rs.normal(size=(100, 26)),
                 columns=list(ascii_letters[26:]))
corr = d.corr()
corr
sns.heatmap(corr)
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask)
```
