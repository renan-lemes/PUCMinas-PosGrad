# Tupla


* Declaramos uma tupla utilizando ()
* Acessamos informações da tupla com []
* Operações como +, in, [:] e * tem a mesma funcionalidade na tupla que da lista.

```python
exemplo_tupla = (1, 2, 'a')
print(exemplo_tupla)
print(type(exemplo_tupla))
```

```python
print(f'Acessando a posição 0: {exemplo_tupla[0]}')
print(f'Acessando a última posição: {exemplo_tupla[-1]}')
```

```python
len(exemplo_tupla)
```

```python
exemplo_tupla[0] = 3
```

```python
# adicionando um valor a tupla - não altera o valor original
exemplo_tupla + ('b', )
```

```python
exemplo_tupla
```

```python
exemplo_tupla[:2]
```

```python
exemplo_tupla[::-1]
```

```python
exemplo_tupla * 2
```

```python
2 in exemplo_tupla
```

**Atenção**

Mesmo a tupla sendo não mutável ela aceita objetos mutáveis.

```python
lista = [1,2,3]
tupla_lista = (2, lista)
print(f'Tupla declarada {tupla_lista}')
lista.append(4)
print(f'Tupla após modificação {tupla_lista}')
```

# Set
* Declaramos uma variável do tipo set com {}
* O tipo set é construído com apenas os valores únicos


Podemos construir um set a partir de uma tupla e lista utilizando a função set.


```python
set_exemplo = {1, 2, 1, 4}
print(set_exemplo)
set_exemplo = {1, 2, 1, 'a'}
print(set_exemplo)
set_exemplo = {1, 2, 1, True, False}
print(set_exemplo)
```

```python
set([1, 1, 3, 4])
```

```python
lista_exemplo = [1, 1, 2, 3, 4]
set_exemplo = set(lista_exemplo)
print(set_exemplo)
lista_exemplo.append(5)
print(set_exemplo)
```

## Adicionando e removendo informações de um set
* Para adicionar informação utilizaremos a função add
* Para remover utilizaremos remove (KeyError) e discard

```python
set_exemplo.add(5) #não aceita lista
print(set_exemplo)
```

```python
set_exemplo.add((6, 7))
print(set_exemplo)
```

```python
set_exemplo.remove(4)
print(set_exemplo)
```

```python
set_exemplo.remove(4)
print(set_exemplo)
```

```python
set_exemplo.discard(4)
print(set_exemplo)
```

```python
set_exemplo.discard(3)
print(set_exemplo)
```

## Operações com set
* União: unifica os dois valores de dois conjuntos em um
* Interseção: cria um conjunto com os valores comuns de conjuntos.
* Diferença: cria um conjunto com os valores do primeiro conjunto que não estão no segundo.
* Diferença simétrica: cria um conjunto com os valores que estão nos dois conjuntos e que não são interseção.

### União

```python
primeiro_set = set([1, 2, 3, 5, 6])
segundo_set = set([5, 6, 7, 8])
primeiro_set.union(segundo_set)
```

```python
primeiro_set | segundo_set
```

### Interseção

```python
primeiro_set.intersection(segundo_set)
```

```python
primeiro_set & segundo_set
```

```python
set_1 = {1, 2}
set_2 = {3, 4}
set_1.intersection(set_2)
```

### Diferença

```python
primeiro_set.difference(segundo_set)
```

```python
primeiro_set - segundo_set
```

### Diferença simétrica

```python
primeiro_set.symmetric_difference(segundo_set)
```

```python
primeiro_set ^ segundo_set
```

```python
print(primeiro_set)
primeiro_set.update(segundo_set)
print(primeiro_set)
primeiro_set &= segundo_set
print(primeiro_set)
primeiro_set -= set([6, 7])
print(primeiro_set)
primeiro_set ^= segundo_set
print(primeiro_set)
```
