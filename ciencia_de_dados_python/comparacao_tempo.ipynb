```python
import time
import numpy as np
```

## Multiplicação escalar

```python
tamanho = 10000000
array = np.random.rand(tamanho, 1)
array2 = np.random.rand(tamanho, 1)
```

```python
resultado = np.zeros((tamanho, 1))

inicio = time.process_time()

for i in range(tamanho):
  resultado = 3 * array[i]

fim = time.process_time()

print(f'Tempo total: {100 * (fim-inicio)} ms')
```

```python
inicio = time.process_time()

resultado = 3 * array

fim = time.process_time()

print(f'Tempo total: {100 * (fim-inicio)} ms')
```

## Multiplicação de vetores

```python
resultado = 0

inicio = time.process_time()

for i in range(tamanho):
  resultado += array[i] * array2[i]

fim = time.process_time()

print(f'Tempo total: {100 * (fim-inicio)} ms')
```

```python
inicio = time.process_time()

resultado = np.dot(array.T, array2)

fim = time.process_time()

print(f'Tempo total: {100 * (fim-inicio)} ms')
```

## Multiplicação elemento-a-elemento de matriz

```python
tamanho = 1000
```

```python
matriz1 = np.random.rand(tamanho, tamanho)
matriz2 = np.random.rand(tamanho, tamanho)

resultado = np.zeros((tamanho, tamanho))
```

```python
inicio = time.process_time()

for i in range(tamanho):
  for j in range(tamanho):
    resultado[i, j] = matriz1[i, j] * matriz2[i, j]
fim = time.process_time()

print(f'Tempo total: {100 * (fim-inicio)} ms')
```

```python
inicio = time.process_time()
resultado = matriz1 * matriz2
fim = time.process_time()

print(f'Tempo total: {100 * (fim-inicio)} ms')
```

## Multiplicação matricial

```python
tamanho = 100
```

```python
matriz1 = np.random.rand(tamanho, tamanho)
matriz2 = np.random.rand(tamanho, tamanho)

resultado = np.zeros((tamanho, tamanho))
```

```python
inicio = time.process_time()

for i in range(tamanho):
  for j in range(tamanho):
    for k in range(tamanho):
      resultado[i, j] += matriz1[i, k] * matriz2[k, j]
fim = time.process_time()

print(f'Tempo total: {100 * (fim-inicio)} ms')
```

```python
inicio = time.process_time()
resultado2 = np.dot(matriz1, matriz2)
fim = time.process_time()

print(f'Tempo total: {100 * (fim-inicio)} ms')
```
