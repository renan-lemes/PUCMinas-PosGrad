# Matriz esparsa

São matrizes que a maior parte do seus valores são zero. Uma forma eficiente de utilizar essas matrizes é salvar informação a respeito apenas dos campos que são diferentes de zero.

```python
import numpy as np
from scipy.sparse import csr_matrix
```

```python
matriz = np.array([[1, 0, 0, 0], [1, 1, 0, 2], [0, 0, 0, 0]])

print('Matriz')
print(matriz)

sparse_matriz= csr_matrix(matriz)
print('---------------')
print('Matriz esparsa')
print(sparse_matriz)
```

```python
sparse_matriz.todense()
```
